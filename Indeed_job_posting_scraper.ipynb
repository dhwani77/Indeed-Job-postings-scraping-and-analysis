Gathering data from the homepage

```python
base_url='https://www.indeed.com/jobs?as_and=&as_phr=&as_any=data+analyst%2C+business+analyst%2C+business+intelligence%2C+project+management%2C+product+management%2C+data+scientist%2C+product+analyst&as_not=&as_ttl=&as_cmp=&jt=fulltime&st=&sr=directhire&salary=&radius=100&l=California&fromage=7&limit=10&sort=&psf=advsrch&from=advancedsearch'
```

```python
#Gathering data from homepage
res=requests.get(base_url)
soup=bs4.BeautifulSoup(res.text,"lxml")
```

```python
soup
```

```python
job_title=[]
job_location=[]
company_name=[]
job_starting_salary=[]
job_list_date=[] 
job_description = []
job_application_link = []


days_to_subtract = 0
length=len(soup.select('.jobtitle'))

for i in range(0,length):
    job_title.append(soup.select('.jobtitle')[i]['title'])
    print(soup.select('.jobtitle')[i]['title'])
    
    job_location.append(soup.select(".location")[i].text)
    print(soup.select(".location")[i].text)
    
    company_name.append(soup.select('.sjcl .company')[i].text.replace('\n',''))
    print(soup.select('.sjcl .company')[i].text.replace('\n',''))

    try:
        ItemId = re.sub(r'(.) (.+)', r'\1', soup.select('.salaryText')[i].text.replace('\n$',''))
    except:
        ItemId = None
    job_starting_salary.append(ItemId)
    print(ItemId)
    
    job_list_date.append(soup.select(".date")[i].text)
    print(soup.select(".date")[i].text)        
    
    job_soup = soup.find(id="resultsCol")
    job_elems = job_soup.find_all('div', class_='jobsearch-SerpJobCard')[i]
    link = job_elems.find('a')['href']
    link = 'http://www.Indeed.com' + link    
    print(link)
    
    time.sleep(10)  # at least 10 seconds pause
    res1=requests.get(link, headers = headers)
    soup1=bs4.BeautifulSoup(res1.text,"lxml")
    
    try:
        applylink = soup1.find_all('a',class_='icl-Button icl-Button--primary icl-Button--lg icl-Button--block')[0]['href']
    except:
        try:
            applylink = soup1.find_all('a',class_='icl-Button icl-Button--primary icl-Button--md icl-Button--block')[0]['href']
        except:
            applylink = link
        
    job_application_link.append(applylink)
    print(applylink)
    
    try:
        description = soup1.find('div', class_ = "jobsearch-JobComponent-description icl-u-xs-mt--md").get_text()
    except:
        description = 'NULL'
        
    job_description.append(description)
    print(description)
```

Gathering data from other pages

```python
for i in range(10,200,10):
    time.sleep(10)  # at least 10 seconds pause
    scrape_url=base_url+'&start={}'.format(i)
    res=requests.get(scrape_url)
    soup=bs4.BeautifulSoup(res.text,"lxml")
    
    length=len(soup.select('.jobtitle'))

    for i in range(0,length):
        job_title.append(soup.select('.jobtitle')[i]['title'])
        job_location.append(soup.select(".location")[i].text)
        company_name.append(soup.select('.sjcl .company')[i].text.replace('\n',''))
        job_list_date.append(soup.select(".date")[i].text)
        
        try:
            ItemId = re.sub(r'(.) (.+)', r'\1', soup.select('.salaryText')[i].text.replace('\n$',''))
        except:
            ItemId = None
        job_starting_salary.append(ItemId)
        
        job_soup = soup.find(id="resultsCol")
        job_elems = job_soup.find_all('div', class_='jobsearch-SerpJobCard')[i]      
        link = job_elems.find('a')['href']
        link = 'http://www.Indeed.com' + link    
    
        time.sleep(10)  # at least 10 seconds pause
        res1=requests.get(link, headers = headers)
        soup1=bs4.BeautifulSoup(res1.text,"lxml")
    
        try:
            applylink = soup1.find_all('a',class_='icl-Button icl-Button--primary icl-Button--lg icl-Button--block')[0]['href']
        except:
            try:
                applylink = soup1.find_all('a',class_='icl-Button icl-Button--primary icl-Button--md icl-Button--block')[0]['href']
            except:
                applylink = link

        job_application_link.append(applylink)

        try:
            description = soup1.find('div', class_ = "jobsearch-JobComponent-description icl-u-xs-mt--md").get_text()
        except:
            description = 'NULL'

        job_description.append(description)   
```

```python
indeed_job_list = pd.DataFrame(list(zip(company_name, job_title,job_location, job_list_date, job_description, job_application_link)), 
               columns =['Company Name', 'Job Title', 'Job Location', 'Job List Date', 'Job Description', 'Job Application Link'])
```

```python
today = datetime.today()
day_subtract = 0
            
for i in range(0,len(job_list_date)):
    try:
        if(indeed_job_list['Job List Date'][i] in ('Today','Just posted')):
            indeed_job_list['Job List Date'][i] = today.strftime("%d/%m/%Y")
        else: 
            day_subtract = re.sub(r'(.) (.+)', r'\1', indeed_job_list['Job List Date'][i])
            indeed_job_list['Job List Date'][i] = (today - timedelta(days = int(day_subtract))).strftime("%d/%m/%Y")
    except:
        indeed_job_list['Job List Date'][i] = today.strftime("%d/%m/%Y")
```
